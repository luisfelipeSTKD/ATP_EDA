```python
# Separando variáveis numéricas de variáveis categóricas

var_num = df2.select_dtypes(include=['float64'])
var_cat = df2.select_dtypes(include=['object', 'bool'])
```

```python
# Para cada uma das variáveis numéricas, deverá apresentar: 
# média;
# mediana;
# variância;
# desvio padrão;
# quartis.

# Média e Mediana
media = pd.DataFrame(var_num.apply(np.mean)).reset_index().T
mediana = pd.DataFrame(var_num.apply(np.median)).reset_index().T

# Variância
variancia = pd.DataFrame(var_num.apply(lambda x: x.max() - x.min())).reset_index().T

# Desvio Padrão
des_padrao = pd.DataFrame(var_num.apply(np.std)).reset_index().T

# Quartis
# quartis = pd.DataFrame(var_num.apply(np.quantile(var_num, 0.25))).T
quartis = pd.DataFrame.quantile(var_num, 0).reset_index().T


# Concatenação
conc_valores = pd.concat([media, mediana, variancia, des_padrao, quartis]).T
cols = 'Média', 'Mediana', 'Variância', 'Desvio Padrão', 'Quartis'
conc_valores.columns = cols # >> "Length mismatch: Expected axis has X elements, new values have Y elements" >> Os valores mudam de acordo com a configuração que coloco
conc_valores
```
